```python
from PIL import Image, ImageDraw, ImageFont, ImageFilter
def split_sentence(sentence):
    words = sentence.split()
    result = []
    current_chunk = ''
    for word in words:
        if len(current_chunk + word) <= 22:
            current_chunk += word + ' '
        else:
            result.append(current_chunk[:-1])
            current_chunk = word + ' '
    result.append(current_chunk[:-1])
    return result


# open image and get size
background = Image.open('Poli_logo_in_animated_style_for_a_chatbot_13e59d50-61b8-42e3-b3ff-855bb349d4b9.png')
background = background.resize((680,680))
image_size = background.size

# create square image with transparent background
margin_size = 100
square_size = (image_size[0]-margin_size, image_size[1]-margin_size)
square = Image.new('RGBA', square_size, (255, 255, 255, 0))

# apply transparency to entire image
alpha = 200
mask = Image.new('L', square_size, 0)
mask_draw = ImageDraw.Draw(mask)
mask_draw.rectangle((0, 0, square_size[0], square_size[1]), fill=alpha)
square.putalpha(mask)

# paste square onto image
square_position = ((image_size[0] - square_size[0]) // 2, (image_size[1] - square_size[1]) // 2)
background.paste(square, square_position, square)

# set font and text
font = ImageFont.truetype("./Roboto/Roboto-Black.ttf", 45)
raw_text = "15. A IA é usada por chatbots para melhorar a experiência do usuário."
text_list = split_sentence(raw_text)
text = '\n'.join(text_list)

# get size of text and calculate position
text_size = font.getsize(max(text_list, key=len))
text_position = ((image_size[0] - text_size[0]) // 2, (image_size[1] - text_size[1]*len(text_list)) // 2)

# create draw object and add text
draw = ImageDraw.Draw(background)
draw.text(text_position, text, fill='black', font=font)

# show final image
display(background)
```
